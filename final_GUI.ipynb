```python
def main(page: ft.Page):
    global start_station, end_station

    # GUI的排版
    page.title = "house price predct"
    page.window_width = 700
    page.window_height = 600
    page.horizontal_alignment = ft.CrossAxisAlignment.CENTER

    data_original = []

    # 按鈕的function
        
    def button_clicked(e):
        data_original.append(a.value)
        data_original.append(b.value)
        data_original.append(c.value)
        data_original.append(d.value)
        data_original.append(e1.value)
        data_original.append(f.value)
        data_original.append(g.value)
        data_original.append(h.value)
        data_original.append(i.value)
        if j.value == "一至四層":
            data_original.extend((0,0,0,0,))
        elif j.value == "五層至十層":
            data_original.extend((1,0,0,0,))
        elif j.value == "十層至十五層":
            data_original.extend((0,1,0,0,))
        elif j.value == "十五層至二十層":
            data_original.extend((0,0,1,0,))
        elif j.value == "二十層以上":
            data_original.extend((0,0,0,1,))
        if k.value == "透天厝":
            data_original.extend((1,0,0))
        elif k.value == "公寓":
            data_original.extend((0,1,0))
        elif k.value == "住宅大樓":
            data_original.extend((0,0,1))
        elif k.value == "華夏":
            data_original.extend((0,0,0))
        if l.value == "鋼骨鋼筋混凝土造":
            data_original.extend((1,0,0,0))
        elif l.value == "鋼筋混凝土造":
            data_original.extend((0,1,0,0))
        elif l.value == "加強磚造":
            data_original.extend((0,0,1,0))
        elif l.value == "磚造":
            data_original.extend((0,0,0,1))
        elif l.value == "木造":
            data_original.extend((0,0,0,0))
        data_original.append(m.value)
        try:
            float_data = list(map(float, data_original))
            print(float_data)
            sample_input = np.array(float_data)
            sample_input_scaled = scaler.transform(sample_input.reshape(1, -1))
            sample_input_tensor = torch.tensor(sample_input_scaled, dtype=torch.float32)
            model.eval()
            with torch.no_grad():
                predicted_price = model(sample_input_tensor)
                print(predicted_price)
                print(f'Predicted Price: ${predicted_price.item() :.2f}')
            text.value = f'Predicted Price/area(m^2): ${predicted_price.item() :.2f}'
        except:
            text.value = '請勿留空'
        page.update()
    a = ft.Dropdown(
            label="是否為特殊用地(農牧用地等)？",
            hint_text="0為否, 1為是",
            options=[
                ft.dropdown.Option(0),
                ft.dropdown.Option(1),
            ],
            autofocus=True,
        )
    b = ft.TextField(label="總樓層數(那棟大樓整棟)")
    c = ft.TextField(label="建物總面積平方公尺")
    d = ft.TextField(label="房數")
    e1 = ft.TextField(label="廳數")
    f = ft.TextField(label="衛浴數")
    
    g = ft.Dropdown(
            label="有無電梯",
            hint_text="0為否, 1為是",
            options=[
                ft.dropdown.Option(0),
                ft.dropdown.Option(1),
            ],
            autofocus=True,
        ) 
    h = ft.TextField(label="土地數")
    i = ft.TextField(label="建物數")
    j = ft.Dropdown(
            label="樓層(你買的那層)",
            hint_text="",
            options=[
                ft.dropdown.Option("一至四層"),
                ft.dropdown.Option("五層至十層"),
                ft.dropdown.Option("十層至十五層"),
                ft.dropdown.Option("十五層至二十層"),
                ft.dropdown.Option("二十層以上"),
            ],
            autofocus=True,
        ) 
    k = ft.Dropdown(
            label="建物型態",
            hint_text="",
            options=[
                ft.dropdown.Option("透天厝"),
                ft.dropdown.Option("公寓"),
                ft.dropdown.Option("住宅大樓"),
                ft.dropdown.Option("華夏"),
            ],
            autofocus=True,
        ) 
    l = ft.Dropdown(
            label="主要建材",
            hint_text="",
            options=[
                ft.dropdown.Option("鋼骨鋼筋混凝土造"),
                ft.dropdown.Option("鋼筋混凝土造"),
                ft.dropdown.Option("加強磚造"),
                ft.dropdown.Option("磚造"),
                ft.dropdown.Option("木造"),
            ],
            autofocus=True,
        ) 
    m = ft.Dropdown(
            label="屬於舊台南市嗎",
            hint_text="0為否, 1為是",
            options=[
                ft.dropdown.Option(0),
                ft.dropdown.Option(1),
            ],
            autofocus=True,
        )  
    hint1 = ft.Text("是非題1為是,0為否")
    hint2 = ft.Text("勿留空")
    
    button = ft.ElevatedButton(text="提交", on_click=button_clicked)
    text = ft.Text("Price:")
    page.add(hint1, hint2, a, b, c, d, e1, f, g, h, i, j, k, l, m, button, text)
   

ft.app(target=main)
```
